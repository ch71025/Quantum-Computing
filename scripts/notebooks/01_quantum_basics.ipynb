## 1. Classical vs Quantum Bits

**Classical Bit:**
- Either 0 OR 1
- Like a light switch: on/off

**Quantum Bit (Qubit):**
- Can be 0 AND 1 simultaneously
- Like a spinning coin: both heads/tails until measured
- Described by: $|\psi\rangle = \alpha|0\rangle + \beta|1\rangle$

## 2. Probability Amplitudes

For a qubit state $|\psi\rangle = \alpha|0\rangle + \beta|1\rangle$:
- $|\alpha|^2$ = probability of measuring 0
- $|\beta|^2$ = probability of measuring 1
- $|\alpha|^2 + |\beta|^2 = 1$ (normalization)

## 3. First Quantum Concept: Superposition

The Hadamard gate creates equal superposition:
$$
H|0\rangle = \frac{|0\rangle + |1\rangle}{\sqrt{2}}
$$
$$
H|1\rangle = \frac{|0\rangle - |1\rangle}{\sqrt{2}}
$$
